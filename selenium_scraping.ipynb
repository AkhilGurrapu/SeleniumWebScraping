## Logic for the "Load more products" button at bottom

```python
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
import csv
from selenium.common.exceptions import TimeoutException
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By

# Open a CSV file to write the product data
with open('products.csv', mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    
    # Write the header row
    writer.writerow(['Category', 'Product'])
    
    while True:
        try:
            # Locate the 'Load More' button by the data-cy attribute
            loadButton = WebDriverWait(driver, 10).until(
                EC.element_to_be_clickable((By.CSS_SELECTOR, "[data-cy='button-label']"))
            )
    
            # Click the button
            loadButton.click()
    
            # Wait for the next set of products to load
            # Here you need to add a suitable wait condition if necessary
            WebDriverWait(driver, 10).until(EC.visibility_of_element_located((By.CLASS_NAME, "line-clamp-2")))
    
            # Find all elements with the class name "line-clamp-2"
            products = driver.find_elements(By.CLASS_NAME, "line-clamp-2")
            product_category = driver.find_elements(By.CSS_SELECTOR, 'p.m-0.cursor-pointer.truncate.text-left.text-sm.font-bold.uppercase.leading-4')
            
    
            # Iterate over each product element and print its text
            for product, category in zip(products, product_category):
                print(f"Product Name: {product.text}, Product Category: {category.text}")
    
        except TimeoutException:
            # Break the loop if the 'Load More' button is not found or not clickable
            print("No more 'Load More' button found, or it's not clickable.")
            break
```
